**Task:**

Imagine you're a tourist in Pakistan, and you want to travel from Islamabad to Karachi. You
want to find the shortest route in terms of the number of cities you need to travel through. The
cities are connected through highways, and your goal is to find the shortest path using BFS.

**Cities and Their Connections (Graph):**

You will use the following cities and their connections for this scenario:
* **Islamabad** is connected to **Rawalpindi, Lahore, and Peshawar.**

* **Rawalpindi** is connected to **Islamabad, Peshawar, and Quetta.**
* **Peshawar** is connected to **Islamabad, Rawalpindi, and Quetta.**
* **Lahore** is connected to **Islamabad, Multan, and Quetta.**
* **Multan** is connected to **Lahore, Karachi, and Quetta.**
* **Quetta** is connected to **Rawalpindi, Peshawar, Multan, and Karachi.**
* **Karachi** is connected to **Multan and Quetta.**

```python
#importing libraries
import networkx as nx
import matplotlib.pyplot as plt

#Graph Definition
cities_graph = {
    'Islamabad': ['Rawalpindi', 'Lahore', 'Peshawar'],
    'Rawalpindi': ['Islamabad', 'Peshawar', 'Quetta'],
    'Peshawar': ['Islamabad', 'Rawalpindi', 'Quetta'],
    'Lahore': ['Islamabad', 'Multan', 'Quetta'],
    'Multan': ['Lahore', 'Karachi', 'Quetta'],
    'Quetta': ['Rawalpindi', 'Peshawar', 'Multan', 'Karachi'],
    'Karachi': ['Multan', 'Quetta']
}

#Creating Graph obj
G = nx.Graph()

#loops to add edges
for city, neighbors in cities_graph.items():
    for neighbor in neighbors:
        G.add_edge(city, neighbor)

# Setting up the plot
plt.figure(figsize=(8, 6))

# positions for all nodes
pos = nx.spring_layout(G)

# Drawing the graph
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold', edge_color='black')

plt.title("Cities and Their Highways")

plt.show()
```

**Code for Shotest Path**

```python
#importing libraries
from collections import deque

#BFS Shortest Path Function
def bfs_shortest_path(graph, start, goal):

#Initializing BFS Queue and visited set
    queue = deque([(start, [start])])
    visited = set()

 # Main BFS Loop
    while queue:
        current_city, path = queue.popleft()

 # Checking if the Goal City is Reached
        if current_city == goal:
            return path

 # Making the Current City as Visited
        visited.add(current_city)

 # Exploring Neighboring Cities
        for neighbor in graph[current_city]:
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor]))

 # Return None if no path is found
    return None

 # Define the graph with cities and their connections
cities_graph = {
    'Islamabad': ['Rawalpindi', 'Lahore', 'Peshawar'],
    'Rawalpindi': ['Islamabad', 'Peshawar', 'Quetta'],
    'Peshawar': ['Islamabad', 'Rawalpindi', 'Quetta'],
    'Lahore': ['Islamabad', 'Multan', 'Quetta'],
    'Multan': ['Lahore', 'Karachi', 'Quetta'],
    'Quetta': ['Rawalpindi', 'Peshawar', 'Multan', 'Karachi'],
    'Karachi': ['Multan', 'Quetta']
}

# Setting Start and Goal Cities
start_city = 'Islamabad'
goal_city = 'Karachi'

# Running the BFS to Find The Shortest Path
shortest_path = bfs_shortest_path(cities_graph, start_city, goal_city)

# Output the Shortest Path
if shortest_path:
    print("Shortest path from", start_city, "to", goal_city, "is:", " -> ".join(shortest_path))
else:
    print("No path found from", start_city, "to", goal_city)
```
